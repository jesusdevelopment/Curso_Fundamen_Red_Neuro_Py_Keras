## Importación de Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Funciones de Activación

### Función Sigmoide

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.title('Función Sigmoide')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.show()
```

### Función Escalonada

#### Usando Where

```python
def step(x):
    return np.where(x >= 0, 1, 0)

x = np.linspace(-10, 10, 100)
y = step(x)

plt.plot(x, y)
plt.title('Función Escalonada')
plt.xlabel('x')
plt.ylabel('step(x)')
plt.show()
```

#### Usando Piecewise

```python
def step_piecewise(x):
    return np.piecewise(x, [x < 0, x >= 0], [0, 1])

x = np.linspace(-10, 10, 100)
y = step_piecewise(x)

plt.plot(x, y)
plt.title('Función Escalonada')
plt.xlabel('x')
plt.ylabel('step_piecewise(x)')
plt.show()
```

### Función Relu

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10, 10, 100)
y = relu(x)

plt.plot(x, y)
plt.title('Función ReLU')
plt.xlabel('x')
plt.ylabel('relu(x)')
plt.show()
```

## Función Tanh

### Función Directa en Numpy

```python
def tangh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 100)
y = tangh(x)

plt.plot(x, y)
plt.title('Función Tanh')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.show()
```

#### Usando exponencial

```python
def tanh(x):
    return (np.exp(2*x) - 1) / (np.exp(2*x) + 1)

x = np.linspace(-10, 10, 100)
y = tanh(x) 

plt.plot(x, y)
plt.title('Función Tanh')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.show()
```

## Funciones de Pérdida

### MSE

```python
def MSE(Y, Y_hat, derivate=False): 
    if derivate: 
        return 2 * (Y_hat - Y)
    else:
        return np.mean((Y_hat - Y) ** 2)
```

```python
real = np.array([0, 0, 1, 1]) 
prediction = np.array([0.9, 0.5, 0.2, 0.0])
print(MSE(real, prediction))
```
